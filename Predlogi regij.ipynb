## Drseče okno

Z drsečim oknom

+ V - velikosti kvadratnega okna  poveča)
+ S - korak okna
+ H,W - širina H in višina W slike
+ N - število oken 

$$
h = \left\lfloor {(H-V)}/{S} \right\rfloor+1 \\
$$
$$
w = \left\lfloor {(W-V)}/{S} \right\rfloor+1 \\
$$
$$
N = h \cdot w
$$


```python
H = I.shape[0]
W = I.shape[1]

velikosti_oken = [60.,80.,100.,120.,140.]
koraki_oken =    [ 6., 8., 10., 12., 14.] # [12.,16.,20.,24.,28.] #

N = 0
for i in range(0,len(velikosti_oken)):
    V = velikosti_oken[i]
    S =koraki_oken[i]
    h = floor((H-V)/S)+1
    w = floor((W-V)/S)+1
    print 'število oken za okno ({0:d},{1:d}): {2:d}'.format(int(V),int(S),int(h*w))
    N += h*w
print 'število vseh oken:{0:d}'.format(int(N))
```

```python
# velikost slike
H = I.shape[0]
W = I.shape[1]

okna = []

for i in range(0,len(velikosti_oken)): 
    V = int(velikosti_oken[i])
    S = int(koraki_oken[i])
    x = arange(0,W-V,S) # seznam pozicij x drsečega okna
    y = arange(0,H-V,S) # seznam pozicij y drsečega okna

    # funkcija meshgrid prejme seznam pozicij in vrne mrežo teh pozicij
    # mesgrid([1,2],[1,2,3]) vrne X=[[1,1,1],[2,2,2]] in Y= [[1,2,3],[1,2,3]] 
    X,Y = meshgrid(x,y) 
    
    # pozicije okna X in Y skupaj z velikostmi sestavimo v wins
    wins = vstack((X.ravel(),Y.ravel(),V*ones(X.ravel().shape[0]),V*ones(X.ravel().shape[0])))
    okna += [wins]


okna = hstack(okna).T
```

```python
def OcenaPrekrivanja(Rc,Rk):
    '''
    Ocena prekrivanja
    
    Funkcija primerja ciljno okno z generiranimi okni in 
    generirana okna oceni z deležem prekrivanja. Delež
    prekrivanja je površina unije deljena s površino preseka,
    med ciljnim in preizkušenim oknom.
    
    Odstrani okna, ki se ne prekrivajo z ciljnim oknom in 
    jih razporedi po vrsti, po oceni.
    
    
    Rc - array(4,)
        Ciljno okno, katerega želimo najti.
    Rk - array(N,4)
        Okna katera smo generirali oziroma preizkusili.
    
    Vrne:
    Rk - array(M,4)
        Okna, ki se prekrivajo z ciljnim oknom. Razporejena po oceni.
    ocena - array(M)
        Ocena oken Rk.
    
    '''
    levo =  where(Rk[:,0]>Rc[0],Rk[:,0],Rc[0])
    vrh =   where(Rk[:,1]>Rc[1],Rk[:,1],Rc[1])
    
    desno = where(Rk[:,0]+Rk[:,2]<Rc[0]+Rc[2],\
                  Rk[:,0]+Rk[:,2],Rc[0]+Rc[2])
    dno =   where(Rk[:,1]+Rk[:,3]<Rc[1]+Rc[3],\
                  Rk[:,1]+Rk[:,3],Rc[1]+Rc[3])
    # izločimo tiste, ki se ne prekrivajo
    M = ((levo < desno)*(vrh < dno))>0 
    Rk = Rk[M,:]
    # izračunamo okna presekov
    Rp = vstack([levo[M],vrh[M],desno[M]-levo[M],dno[M]-vrh[M]]).T
    
    presek = Rp[:,2]*Rp[:,3]
    unija =  Rk[:,2]*Rk[:,3]+Rc[2]*Rc[3] - presek
    
    ocena = presek/unija
    inds = flipud(argsort(ocena))
    
    return Rk[inds,:],ocena[inds],inds

# poiscemo okno, ki se najbolje prekriva z posameznim ciljnim oknom
R_najb = []
for n in range(0,3):
    R_kandidati,ocena,_ = OcenaPrekrivanja(R_cilj[n,:],okna) 

    R_najb += [R_kandidati[0,:]]
    
R_najb = vstack(R_najb)

# izris 

fig = figure()
imshow(I)

for n in range(0,3):
    fig.axes[0].add_artist(Rectangle(R_cilj[n,:2],R_cilj[n,2],R_cilj[n,3],\
                                     fill=False,ec = (0.,1.,0.),lw=2.))
for n in range(0,R_najb.shape[0]):
    fig.axes[0].add_artist(Rectangle(R_najb[n,:2],R_najb[n,2],R_najb[n,3],\
                                     fill=False,ec = (1.,0.,0.),lw=2.))
    
```

## Metoda selektivnega iskanja s segmentacijo gradi na naslednjih principih

### Zajemanje različnih velikosti
Predmeti v sliki so lahko različnih velikosti. Segmentacija mora biti temu prilagojena. Zato uporabimo hierarhično - drevesno združevanje segmentov. Večji segmenti so sestavljeni z združevanjem manjših. 
### Raznolikost
Težko bomo našli optimalno strategijo segmentacije, ki se bo obnesla v vseh slikah. Zato uporabimo različne segmentacije in različne metrike združevanja. Ta bodo generirala večje število segmentov - regij, kar poveča možnosti, da se iskan predmet nahaja v eni zmed regij. Ker se bodo v različnih strategijah segmentacije pojavljali isti ali podobni segmenti, __iz končnega seznama predlaganih regij odstranimo podvojene__.
### Hitrost
Z iskanjem regij poskušamo omejiti za zahtevno nalogo prepoznave oziroma detekcije predmetov. Iskanje regij zato ne sme postati časovno najzahtevnejša operacija.
    

## Drevesno združevanje regij

### Začetna segmentacija 
Za začetno segmentacijo slike uporabimo izbrano metodo segmentacije, ki bo sliko razdelila v majhne segmente. Želimo, da bi vsak segment vseboval le en osnovni predmet. Avtorji uporabijo metodo segmentacije z grafi:

+ P. F. Felzenszwalb and D. P. Huttenlocher. Efficient Graph-Based Image Segmentation. IJCV, 59:167–181, 2004.

### Drevesno združevanje segmentov

Drevesno združevanje opravimo od spodaj navzgor. Za združevanje uporabimo požrešen algoritem. Najprej izračunamo __podobnost__ med vsakim parom __sosednjih regij__. Nato __najpodobnejši regiji odstranimo__ jih združimo in __dodamo kot novo regijo__. __Izračune podobnosti posodobimo__ z izračunom podobnosti nove regije in njenih sosednjih regij.  Nadaljujemo, dokler nam ne ostane ena sama regija celotne slike.


### Mera podobnosti

Za mere podobnosti uporabimo raznolike dopolnjujoče mere in značilke. Za hitro iračunanje je pomembno, da lahko značilke novega segmenta z izračunamo iz značilk združenih segmentov. Avtorji predlagajo 4 mere:

----

+ __mera barve__ meri podobnost dveh regij s primerjanjem normaliziranih barvnih histogramov

$$
s_{colour}(r_i,r_j) = \sum_{h=1}^H{min(c_i(h),c_j(h))} \ \ \leftarrow ocena
$$
$$
    c_k(h) = {{size(r_i)c_i(h)+size(r_j)c_j(h)} \over {size(r_i)+size(r_j)}} \ \ \leftarrow histogram\ novega\ segmenta
$$

----

+ __mera teksture__ meri podobnost dveh regij s primerjanjem SIFT značilk, oziroma histogramov orientiranih gradientov (HoG)
$$
s_{texture}(r_i,r_j) = \sum_{h=1}^H{min(t_i(h),t_j(h))} \ \ \leftarrow ocena
$$
$$
    t_k(h) = {{size(r_i)t_i(h)+size(r_j)t_j(h)} \over {size(r_i)+size(r_j)}} \ \ \leftarrow histogram\ novega\ segmenta
$$

----
+ __mera velikosti__ vzpodbuja združevanje majhnih regij, da so opazovane regije vedno podobnih velikosti
$$
s_{size}(r_i,r_j) = {{size(r_i)+size(r_j)} \over {size(im)}} \ \ \leftarrow ocena
$$
$$
    size(r_k) = size(r_i)+size(r_j) \ \ \leftarrow velikost\ novega\ segmenta
$$

----

+ __mera dopolnjevanja__ meri kako dobro se regiji dopolnjujeta
$$
s_{fill}(r_i,r_j) = 1-{{size(BB(r_i,r_j))-size(r_i)-size(r_j)} \over {size(img)}}
$$

----

Skupna mera podobnosti med regijama $r_i$ in $r_j$ se izračuna kot

$$
s(r_i,r_j) = a_1 s_{colour}(r,i,_rj)+a_2 s_{texture}(r,i,_rj)+a_3 s_{size}(r,i,_rj)+a_4 s_{fill}(r,i,_rj)
$$

kjer je $a_n=1$, če neko mero želimo uporabiti, ali $0$ če je ne.

### Doseganje raznolikosti

Poleg uporabe različnih mer podobnosti, lahko uporabimo še dodatne strategije za doseganje raznolikosti. Uporabimo lahko različne barvne prostore in različne parametre segmentacije. 

Avtorji preizkusijo uporabo različnih barvnih prostorov oz barvnih kanalov:
    
+ HSV 
+ I (sivinska slika)
+ RGB 
+ rgI (normaliziran rdeč in zelen kanal + intenziteta)
+ Lab
+ H (samo Hue kanal iz HSV)
+ rgb (normaliziran RGB)
+ C ("opponent colour space")



Avtorji preizkusijo vse različne kombinacije mer podobnosti. In preizkusijo spreminjanje parametra metode segmentacije - prag.

## Category-Independent Object Proposals
I. Endres and D. Hoiem

Procesiranje traja nekaj minut.

Algoritem sliko najprej hierarhično segmentira - rezultat so segmentacije z različno velikimi segmenti. Iz segmentiranih slik poišče semena - malce večje regije. Za vsako seme poišče binarno segmentacijo predmet-ozadje. Tako segmentirano regijo predmeta doda med predloge. 



```python
imshow(imread('CIProposals_img1.png'));xticks([]);yticks([]);title('Slika iz clanka "Category Idependent Object Proposals"');
```

Vse predloge uredi po vrsti tako, da so višje v seznamum boljše in bolj raznovrstne regije. Tako lahko za samo klasifikacijo pogledamo samo nekaj boljše uvrščenih predlaganih regij. 

```python
D = scipy.io.matlab.loadmat('CIOProposalsWRankResults.mat')

# seg - segmentirana slika, vsak piksel ima vrednost segmenta kateremu spada
seg = D['superpixels']
# prop - rangirani predlogi - vsak predlog vsebuje seznam vrednosti segmentov, iz katerih je sestavljen
prop = D['proposals'][:,:]

# za predloge generiramo regije
okna_PRanked = numpy.zeros((prop.shape[0],4))
for n in range(0,prop.shape[0]):
    
    # in1d poišče mesta iz prvega seznama, katerih vrednost je v drugem seznamu
    # sliko seg je zato potrebno linearizirati, rezultat pa nato preoblikovati nazaj v obliko slike
    M = numpy.in1d(seg.ravel(),D['proposals'][n,0]) 
    M = M.reshape(seg.shape)
    
    # seštejemo vrednosti po stolpcih in vrsticah
    M_cols = M.sum(0)
    M_rows = M.sum(1)
    # poiščemo indekse stolpcov in vrstic, kjer je seštevek bil pozitiven
    ind_cols = numpy.where(M_cols>0)[0]
    ind_rows = numpy.where(M_rows>0)[0]
    # okno okrog predlagane regije je sestavljeno iz skrajnic indeksov stolpca in vrstice 
    okna_PRanked[n,:] = ind_cols.min(),ind_rows.min(),ind_cols.max(),ind_rows.max()

# obliko (levo,zgoraj,desno,spodaj) spremenimo v (x,y,w,h)
okna_PRanked[:,2:] -= okna_PRanked[:,:2]

print 'število oken:', okna_PRanked.shape[0]
```

```python
R_najb_PRanked = []
for n in range(0,3):
    R_kandidati,ocena,inds = OcenaPrekrivanja(R_cilj[n,:],okna_PRanked) 
    R_najb_PRanked += [R_kandidati[0,:]]
    
R_najb_PRanked = vstack(R_najb_PRanked)

fig = figure()
imshow(I)

for n in range(0,3):
    fig.axes[0].add_artist(Rectangle(R_cilj[n,:2],R_cilj[n,2],R_cilj[n,3],\
                                     fill=False,ec = (0.,1.,0.),lw=2.))
for n in range(0,R_najb_PRanked.shape[0]):
    fig.axes[0].add_artist(Rectangle(R_najb_PRanked[n,:2],R_najb_PRanked[n,2],R_najb_PRanked[n,3],\
                                     fill=False,ec = (1.,0.,0.),lw=2.))
    
xticks([]);yticks([]);
```

## Primov algoritem
Primov algoritem poišče drevo z maksimalno oceno povezav, ki povezuje vsa vozlišča grafa. Iz povezav vseh točk v drevesu v točke izven drevesa vzame povezavo z maksimalno oceno. Izbrano povezavo in njeno točko doda v drevo. Postopek ponavlja dokler na doda vseh točk.

## Naključen primov algoritem
Algoritem v drevo doda naključno povezavo. Ocene povezav interpretira kot verjetnosti. Tako ima povezava z višjo oceno samo večjo verjetnost, da bo dodana v drevo. Algoritem ima tudi dodatno kriterijsko funkcijo prekinitve. Ta preprečuje, da bi algoritem v drevo dodal vse segmente slike v drevo.
