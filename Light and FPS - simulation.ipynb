
Lets replicate this phenomena with a simple calculation - a simulation of the phenomena if you will.

First we create the underling signal of the simulation. The sampling frequency of the signal need to be high enough to accuratly represent all of the simulated signals. By carefully selecting this frequency, so that it is divisible by the frequency of the signals we wish to replicate, we also avoid any nasty rounding problems (sampling errors). 

```python
# frame rates 30Hz,25Hz
# light/electrical frequency 60Hz,50Hz

T = 1. # lenght of simulated recording in seconds
F_sampling = 3000 
# a 300Hz sampling frequency of the simulated signal, so that it is divisible by 25,30,50,60
# according to niquist, this must be at least 2x the largest frequency we wish to simulate
X = np.arange(0,T*F_sampling)/(F_sampling)
```

We simulate two types of light - one that is powered with 60Hz and the other with 50Hz grid power frequency. We show a small segment of the signal.

```python
# the simulated blinking of light, as a  sine wave
L_60 = sin(X*2*pi*60.)# 60 Hz
L_50 = sin(X*2*pi*50.)# 50 Hz

plot(X,L_60,'-',label='60Hz')
plot(X,L_50,'-',label='50Hz')
legend(loc='best')
xlim(0,0.1)
print ''
```

Now we prepare the camera sampling functions. We will represent these with a matrices. The rows of the matrix represet the time component and the columns represent the frames of the video. The values of the matrix are mostly 0. We set them to 1 only where we ant to expose any given frame.

The exposure is given in samples of the underlying simulation. For 25 FPS, the maximum exposure is F_sampling/25, for 30 FPS the maximum exposure is F_sampling/30.

```python

'''
exposure - number of samples to sum
fps - target camera framerate
T - total time of video
sampling - the sampling rate of the simulation
'''
def CreateCameraSamplingMatrix(exposure,fps,T,sampling):
    C = np.zeros((T*sampling,T*fps))
    x_step = sampling/fps
    x = 0
    for f in range(0,int(T*fps)):
        C[x:x+exposure,f] = 1.
        x += x_step
        
    return C

# sampling for 30 Hz camera, each column represents a single frame
C_30 = CreateCameraSamplingMatrix(exposure = F_sampling/30,
                                  fps = 30,
                                  T = T,
                                  sampling = F_sampling)
C_25 = CreateCameraSamplingMatrix(exposure = F_sampling/25,
                                  fps = 25,
                                  T = T,
                                  sampling = F_sampling)
  

subplot(211)
plot(X,L_50**2)
plot(X,C_30)
xlim(0,0.1)
ylim(0,1.1)
subplot(212)
plot(X,L_60**2)
plot(X,C_25)
xlim(0,0.1)
ylim(0,1.1)
print ''
```

We are interested in the power of the light, so we square it. We then multiply it with the camera matrices. This calculates the amount of light (power of light), captured by each frame, similar to single camera pixel.

```python

# we sample the power of the blinking light, so we square it
C_30_L_60 = (L_60**2).dot(C_30)
C_30_L_50 = (L_50**2).dot(C_30)
C_25_L_60 = (L_60**2).dot(C_25)
C_25_L_50 = (L_50**2).dot(C_25)


subplot(211)
plot(C_30_L_60,label='FPS 30, light 60 Hz')
plot(C_30_L_50,label='FPS 30, light 50 Hz')
legend(loc='upper left')
ylim(40,80)
subplot(212)
plot(C_25_L_60,label='FPS 25, light 60 Hz')
plot(C_25_L_50,label='FPS 25, light 50 Hz')
legend(loc='upper left')
ylim(50,80)
print ''
```

At full exposure 50Hz light will flicker in 30FPS, and 60Hz will flicker in 25FPS. Note that signals here are not scaled equally.

Lets try to change the exposure to reduce this affect.

```python
print 'sampling rates:\n {:d} samples for 30 FPS\n{:d} samples for 25 FPS'.format(F_sampling/50,F_sampling/30)

C_30 = CreateCameraSamplingMatrix(exposure = F_sampling/50,
                                  fps = 30,
                                  T = T,
                                  sampling = F_sampling)
C_25 = CreateCameraSamplingMatrix(exposure = F_sampling/30,
                                  fps = 25,
                                  T = T,
                                  sampling = F_sampling)


C_30_L_60 = (L_60**2).dot(C_30)
C_30_L_50 = (L_50**2).dot(C_30)
C_25_L_60 = (L_60**2).dot(C_25)
C_25_L_50 = (L_50**2).dot(C_25)


subplot(211)
plot(C_30_L_60,label='FPS 30, light 60 Hz')
plot(C_30_L_50,label='FPS 30, light 50 Hz')
legend(loc='upper left')
ylim(25,35)
subplot(212)
plot(C_25_L_60,label='FPS 25, light 60 Hz')
plot(C_25_L_50,label='FPS 25, light 50 Hz')
legend(loc='upper left')
ylim(45,55)
print ''
```

With the correct exposure flickering seems to have been at least reduced if not eliminated. However, this means we most likely underexposed our video. The result might be a very grainy, noisy picture. However under some circumstainces this might be preferable. In post processing removing image noise is easyer than correcting for flickering lights.
